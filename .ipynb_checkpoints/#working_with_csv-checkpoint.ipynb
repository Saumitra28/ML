Index_Col parameter

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['Sno','Name','Release_year','Rating','Votes','Genres'],index_col='Sno')
```

Usecols parameter

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['Sno','Name','Release_year','Rating','Votes','Genres'],index_col='Sno',usecols=['Sno','Name','Rating'])
```

Converters:
Suppose if want to replace the Royal challangers Banglore by RCB. So how ?

```python
def rename(name):
    if name=="Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```
